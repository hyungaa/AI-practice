```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 랜덤 데이터 생성 (2차원 데이터)
np.random.seed(0)
X = np.random.randn(100, 2)  # (100개 샘플, 2차원 데이터)

# 데이터 시각화
plt.scatter(X[:, 0], X[:, 1], alpha=0.5)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("원본 데이터 분포")
plt.show()

# PCA 모델 생성 (1개 주성분 유지)
pca = PCA(n_components=1)

# 차원 축소 수행
X_pca = pca.fit_transform(X)

# 결과 출력
print("PCA 변환 후 데이터 형태:", X_pca.shape)  # (100, 1)

# 변환된 데이터 다시 복원 (시각화를 위해)
X_recovered = pca.inverse_transform(X_pca)

# 원본 데이터 & 변환된 데이터 비교
plt.scatter(X[:, 0], X[:, 1], alpha=0.5, label="Original Data")
plt.scatter(X_recovered[:, 0], X_recovered[:, 1], alpha=0.8, label="PCA Reduced", color="r")
plt.legend()
plt.title("PCA 차원 축소 결과")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 목표 함수 f(x) = x^2 + 4x + 5
def f(x):
    return x**2 + 4*x + 5

# 미분 함수 f'(x) = 2x + 4
def df(x):
    return 2*x + 4

# 학습률 설정
learning_rate = 0.1

# 초기값 설정
x = 5  # 초기 x 값
history = [x]  # 경로 기록

# 경사 하강법 반복 수행
for _ in range(20):
    x = x - learning_rate * df(x)  # x 업데이트
    history.append(x)

# 결과 출력
print("최소값에 수렴하는 x:", x)
print("최소값:", f(x))

# 시각화
x_vals = np.linspace(-6, 2, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x)")
plt.scatter(history, [f(h) for h in history], color='red')  # 경로 표시
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("경사 하강법을 이용한 최적화")
plt.legend()
plt.show()
```

```python
# 목표 함수 f(x, y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# 편미분 함수
def df_dx(x, y):
    return 2*x  # x에 대한 미분

def df_dy(x, y):
    return 2*y  # y에 대한 미분

# 학습률 설정
learning_rate = 0.1

# 초기값 설정
x, y = 4, -3  # 초기값
history = [(x, y)]  # 경로 기록

# 경사 하강법 반복 수행
for _ in range(20):
    x = x - learning_rate * df_dx(x, y)
    y = y - learning_rate * df_dy(x, y)
    history.append((x, y))

# 결과 출력
print("최소값에 수렴하는 (x, y):", (x, y))
print("최소값:", f(x, y))

# 시각화 (3D 그래프)
from mpl_toolkits.mplot3d import Axes3D

x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter(*zip(*history), f(x, y), color='r', s=50, label="경로")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("f(X, Y)")
ax.set_title("다변수 함수에서의 경사 하강법")
plt.legend()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 칼만 필터 객체 생성
kalman = cv2.KalmanFilter(4, 2)

# 상태 전이 행렬 A
kalman.transitionMatrix = np.array([[1, 0, 1, 0], [0, 1, 0, 1], [0, 0, 1, 0], [0, 0, 0, 1]], dtype=np.float32)

# 측정 행렬 H
kalman.measurementMatrix = np.array([[1, 0, 0, 0], [0, 1, 0, 0]], dtype=np.float32)

# 프로세스 노이즈와 측정 노이즈 설정 (조정 가능)
kalman.processNoiseCov = np.eye(4, dtype=np.float32) * 0.1
kalman.measurementNoiseCov = np.eye(2, dtype=np.float32) * 1.0


# 랜덤한 움직임 생성
n_samples = 50  # 샘플 개수
actual_positions = np.cumsum(np.random.randn(n_samples, 2) * 5, axis=0)  # 실제 객체 위치
measured_positions = actual_positions + np.random.randn(n_samples, 2) * 3  # 측정된 객체 위치 (노이즈 포함)

# 초기 상태 설정 (측정값의 첫 번째 값으로 설정)
# statePre는 1차원 배열이어야 합니다.
kalman.statePre = np.array([measured_positions[0][0], measured_positions[0][1], 0, 0], dtype=np.float32).reshape(4, 1) #수정된 부분
kalman.statePost = np.array([measured_positions[0][0], measured_positions[0][1], 0, 0], dtype=np.float32).reshape(4, 1) #수정된 부분


# 칼만 필터 적용
filtered_positions = []  # 추정된 객체 위치 저장
for pos in measured_positions:
    measured = np.array(pos, dtype=np.float32).reshape(2, 1)  # 측정값을 칼만 필터 입력 형태로 변환
    kalman.correct(measured)  # 측정값 업데이트
    predicted = kalman.predict()  # 다음 상태 예측
    filtered_positions.append(predicted[:2, 0])  # 예측된 위치 (x, y)를 저장.  [:2, 0]으로 수정

# 결과 시각화
plt.figure(figsize=(8, 6))
plt.plot(actual_positions[:, 0], actual_positions[:, 1], label="Actual Position", color="blue")  # 실제 위치
plt.scatter(measured_positions[:, 0], measured_positions[:, 1], label="Measured Position (Noisy)", color="red", alpha=0.5)  # 측정 위치

filtered_positions_np = np.array(filtered_positions)  # NumPy 배열로 변환
plt.plot(filtered_positions_np[:, 0], filtered_positions_np[:, 1], label="Kalman Filter Output", color="green")  # 칼만 필터 추정 위치

plt.legend()
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("Kalman Filter Object Tracking")
plt.grid(True)

plt.xlim(measured_positions[:, 0].min() - 5, measured_positions[:, 0].max() + 5)  # x축 범위 조정
plt.ylim(measured_positions[:, 1].min() - 5, measured_positions[:, 1].max() + 5)  # y축 범위 조정

plt.show()
```

```python
import numpy as np

# 행렬 A 정의
A = np.array([[2, 1],
              [1, 2]])

# 고유값과 고유벡터 계산
eigenvalues, eigenvectors = np.linalg.eig(A)

print("고유값:\n", eigenvalues)
print("고유벡터:\n", eigenvectors)

# 고유벡터 행렬 P 생성
P = eigenvectors

# 대각 행렬 D 생성
D = np.diag(eigenvalues)

# P의 역행렬 계산
P_inv = np.linalg.inv(P)

# 대각화 검증: A = P D P^-1
A_reconstructed = P @ D @ P_inv

print("재구성된 행렬 A:\n", A_reconstructed)
```

```python
import cv2
import matplotlib.pyplot as plt

# 흑백 이미지 불러오기
image = cv2.imread("1.png", cv2.IMREAD_GRAYSCALE)
image = np.array(image, dtype=np.float32)

# SVD 분해
U, Sigma, Vt = np.linalg.svd(image, full_matrices=False)

# k개의 특잇값만 유지하여 압축
k = 50  # 압축 수준 (작을수록 압축률 증가)
Sigma_k = np.diag(Sigma[:k])
U_k = U[:, :k]
Vt_k = Vt[:k, :]

# 근사 이미지 복원
compressed_image = U_k @ Sigma_k @ Vt_k

# 이미지 출력
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(compressed_image, cmap="gray")
plt.title(f"SVD compression Image (k={k})")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# 최적화할 함수 정의 (목적 함수)
def objective_function(x):
    return x**2 - 4*x + 4  # 간단한 2차 함수

# 초기값 설정
x0 = np.array([0.0])

# 최적화 수행 (Scipy의 minimize 함수 사용)
result = minimize(objective_function, x0)

print("최적값 x:", result.x)
print("최적화된 함수 값:", result.fun)

# 함수 그래프 시각화
x_vals = np.linspace(-2, 6, 100)
y_vals = objective_function(x_vals)

plt.plot(x_vals, y_vals, label="Objective Function")
plt.scatter(result.x, result.fun, color='red', label="Optimal Point")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Optimization problem solution example")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 최적화할 함수 및 그 미분 함수 정의
def function(x):
    return x**2 - 4*x + 4  # 목적 함수

def gradient(x):
    return 2*x - 4  # 함수의 기울기

# 경사 하강법 구현
def gradient_descent(start_x, learning_rate=0.1, iterations=20):
    x = start_x
    history = [x]  # 경로 저장
    for _ in range(iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        history.append(x)
    return history

# 초기값 설정 및 실행
start_x = 6
learning_rate = 0.1
iterations = 20
path = gradient_descent(start_x, learning_rate, iterations)

# 함수 그래프 시각화
x_vals = np.linspace(-1, 7, 100)
y_vals = function(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^2 - 4x + 4")
plt.scatter(path, [function(x) for x in path], color='red', label="Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Optimizatoin using gradient descent method")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])  # y = 2x

# 비용 함수 정의 (MSE)
def cost_function(m, b):
    return np.mean((y - (m*X + b))**2)

# 비용 함수의 기울기 계산
def gradients(m, b):
    dm = -2 * np.mean(X * (y - (m*X + b)))
    db = -2 * np.mean(y - (m*X + b))
    return dm, db

# 경사 하강법 구현
def gradient_descent(m, b, learning_rate=0.1, iterations=100):
    history = []
    for _ in range(iterations):
        dm, db = gradients(m, b)
        m -= learning_rate * dm
        b -= learning_rate * db
        history.append((m, b, cost_function(m, b)))
    return m, b, history

# 초기값 설정 및 실행
m_init, b_init = 0, 0
learning_rate = 0.01
iterations = 100
m_final, b_final, history = gradient_descent(m_init, b_init, learning_rate, iterations)

# 최적화된 직선 출력
plt.scatter(X, y, label="Data")
plt.plot(X, m_final*X + b_final, color='red', label=f"Optimized Line (y={m_final:.2f}x + {b_final:.2f})")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear regression optimization using Gradient descent")
plt.show()

print(f"최적 기울기: {m_final:.4f}, 최적 절편: {b_final:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 최적화할 함수 및 기울기 정의
def function(x):
    return x**2 - 4*x + 4  # 목적 함수

def gradient(x):
    return 2*x - 4  # 함수의 기울기

# SGD 구현
def stochastic_gradient_descent(start_x, learning_rate=0.1, iterations=20):
    x = start_x
    history = [x]  # 경로 저장
    for _ in range(iterations):
        grad = gradient(x)  # 하나의 샘플에서 기울기 계산
        x = x - learning_rate * grad  # SGD 업데이트
        history.append(x)
    return history

# 초기값 설정 및 실행
start_x = 6
learning_rate = 0.1
iterations = 20
path = stochastic_gradient_descent(start_x, learning_rate, iterations)

# 함수 그래프 시각화
x_vals = np.linspace(-1, 7, 100)
y_vals = function(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^2 - 4x + 4")
plt.scatter(path, [function(x) for x in path], color='red', label="SGD Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Optimization using SGD")
plt.show()
```
